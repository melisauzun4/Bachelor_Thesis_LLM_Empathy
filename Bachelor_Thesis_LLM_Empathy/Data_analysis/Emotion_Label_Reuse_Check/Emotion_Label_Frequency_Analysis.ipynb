```python
import pandas as pd
import re

df = pd.read_csv("all_generated_responses.csv")

df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

print(df.shape)
print(df.columns.tolist())
display(df.head())

def strict_label_reused(response, label):
    if pd.isna(response) or pd.isna(label):
        return 0

    response = str(response).lower()
    label = str(label).lower()

    pattern = r"\b" + re.escape(label) + r"\b"
    return int(bool(re.search(pattern, response)))

df["strict_label_reused"] = df.apply(
    lambda row: strict_label_reused(row["response"], row["emotion_label"]),
    axis=1
)

overall_summary = (
    df.groupby("explicitness")["strict_label_reused"]
    .agg(
        reused_count="sum",
        total="count",
        reuse_rate="mean"
    )
    .reset_index()
)

overall_summary["reuse_percentage"] = overall_summary["reuse_rate"] * 100

model_summary = (
    df.groupby(["model", "explicitness"])["strict_label_reused"]
    .agg(
        reused_count="sum",
        total="count",
        reuse_rate="mean"
    )
    .reset_index()
)

model_summary["reuse_percentage"] = model_summary["reuse_rate"] * 100

print("Overall implicit vs explicit:")
display(overall_summary)

print("Model × explicitness:")
display(model_summary)

df.to_csv("strict_emotion_label_reuse_checked_responses.csv", index=False)
overall_summary.to_csv("strict_emotion_label_reuse_overall_summary.csv", index=False)
model_summary.to_csv("strict_emotion_label_reuse_model_summary.csv", index=False)

print("Saved files.")
```
